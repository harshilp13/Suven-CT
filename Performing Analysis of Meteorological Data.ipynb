```python
sns.set(rc = {'figure.figsize':(15, 5)})
```

```python
data['Apparent Temperature (C)'].plot(color='black');
```

```python
cols_plot = ['Apparent Temperature (C)', 'Humidity']
```

```python
axes = data[cols_plot].plot(marker='.', alpha=0.5, linestyle='None', figsize=(15, 16), subplots=True)
for ax in axes:
    ax.set_ylabel('Hourly Totals')
```

```python
data_columns = ['Temperature (C)', 'Apparent Temperature (C)', 'Humidity']
```

```python
weather_monthly_mean = data[data_columns].resample('MS').mean()
weather_monthly_mean.tail(20)
```

```python
import matplotlib.dates as mdates
```

```python
fig, ax = plt.subplots()

ax.plot(weather_monthly_mean['Apparent Temperature (C)'], color='red', label='Apparent Temperature')

ax.plot(weather_monthly_mean['Humidity'], color='blue', label='Humidity')

ax.xaxis.set_major_locator(mdates.YearLocator())
ax.legend()
ax.set_ylabel('Monthly Total Temperature');
```

```python
weather_monthly_mean.plot.line()
```

```python
start, end = '2006-04', '2016-04'

fig, ax = plt.subplots()
ax.plot(data.loc[start:end, 'Apparent Temperature (C)'],marker='.', linestyle='-', linewidth=0.5, label='Monthly Apparent temperature')
ax.plot(data.loc[start:end, 'Humidity'],marker='o', markersize=8, linestyle='-', label='Monthly Humidity of April Resample')
ax.set_ylabel('Temperature')
ax.legend();
```
